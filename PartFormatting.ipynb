## Add Node and Edge Functions

```python
def add_node(ova_str, type, text, nodeID, x, y, schemeID, visible):
    ova_str['AIF']['nodes'].append({"nodeID": nodeID, "text": text, "type": type})
    ova_str['OVA']['nodes'].append({"nodeID": nodeID,
            "visible": visible,
            "x": x,
            "y": y,
            "timestamp": ""})
    if type != "I":
        ova_str['AIF']['schemefulfillments'].append({
                "nodeID": nodeID,
                "schemeID": schemeID
            })
        
    return ova_str
    
def add_l_node(ova_str, type, text, speaker, speakerID, nodeID, x, y, start, visible):
    if text[-1] == '?' or text[-1] == "'":
        ova_str['AIF']['nodes'].append({"nodeID": nodeID, "text": speaker + ": " + text[0:-1], "type": type})
    else:
        ova_str['AIF']['nodes'].append({"nodeID": nodeID, "text": speaker + ": " + text, "type": type})
    ova_str['AIF']['locutions'].append({"nodeID": nodeID,
                "personID": speakerID,
                "start": start,
                "end": None})
    ova_str['OVA']['nodes'].append({"nodeID": nodeID,
                "visible": visible,
                "x": x,
                "y": y,
                "timestamp": ""})
    #Add invisible nodes for each L node, with personId as 0
    return ova_str

def add_edge(ova_str, edgeID, fromID, toID, visible):
    ova_str['AIF']['edges'].append({
                "edgeID": edgeID,
                "fromID": fromID,
                "toID": toID
            })
    
    ova_str['OVA']['edges'].append({
                "fromID": fromID,
                "toID": toID,
                "visible": visible
            })
    
    return ova_str

def add_participant(participant, pID):
    if participant.find(" ") != -1:
        name_array = participant.split(" ", 1)
        fName = name_array[0]
        sName = name_array[1]
    else:
        fName = participant
        sName = ""
    ova_str['AIF']['participants'].append({
                "participantID": pID,
                "firstname": fName,
                "surname": sName
            })
    return ova_str
```

```python
def generate_random_sequence():
    num_list = ""
    while len(num_list) < 19:
        ran_int = random.randint(0,9)
        num_list = num_list + str(ran_int)
    return num_list
        


def create_node_id(node_counter, num_list):
    node_counter = node_counter+1
    nodeID = str(node_counter) + "_" + num_list
    return nodeID, node_counter



def create_edge_id(edge_counter):
    edgeID = edge_counter+1
    edge_counter = edge_counter+1
    return edgeID, edge_counter

def create_participant_id(participant_counter):
    participantID = participant_counter+1
    participant_counter = participant_counter+1
    return participantID, participant_counter
```

```python
def next_timestamp(timestamp):
    new_timestamp = timestamp[:7] + str(int(timestamp[-1]) + 1) 
    return new_timestamp
```
