```python
import pandas as pd
import numpy as np
import random
```

```python
df  = pd.read_csv('formatted_rationale.csv')
data = df.to_numpy()
data
```

```python
questionDict = dict()
personDict = dict()
chosenRationale = dict()
```

```python
totalQuestions = len(df['0'].unique())
totalPeople = len(df['1'].unique())
totalRationale = len(data)

questionsProportion = (totalQuestions/totalRationale)*100
peopleProportion = (totalPeople/totalRationale)*100

while len(chosenRationale) < (int(totalRationale/100)):
# for i in range(0,int(totalRationale/100)):
    tooMany = False
    
    randIndex = random.randrange(totalRationale)
    
#     append 1 to number of times this question has been chosen
    if data[randIndex,1] in questionDict:
        if ((questionDict[(data[randIndex,1])]/int(totalRationale/100))*100) > questionsProportion:
            tooMany = True
        else:
            questionDict[(data[randIndex,1])] = int(questionDict[(data[randIndex,1])])+1
    else:
        questionDict[(data[randIndex,1])] = 1
#      (   questions so far    /   742   )  >=    10%
#     if ((questionDict[(data[randIndex,1])]/int(totalRationale/100))*100) >= questionsProportion:
#         tooMany = True
        
#     append 1 to number of times this person has been chosen
    if data[randIndex,2] in personDict:
        if (personDict[(data[randIndex,2])]/int(totalRationale/100))*100 > peopleProportion:
            tooMany = True
        else:
            personDict[(data[randIndex,2])] = int(personDict[(data[randIndex,2])])+1
    else:
        personDict[(data[randIndex,2])] = 1
#     if (personDict[(data[randIndex,2])]/int(totalRationale/100))*100 >= peopleProportion:
#         tooMany = True

    if tooMany == False and randIndex not in chosenRationale:
        chosenRationale[randIndex] = data[randIndex]
#     else:
#         i = i - 1
#     print(i)
```
